```python

import pandas as pd
df = pd.read_csv('/content/titanic (3).csv')
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Age'].fillna(df['Age'].mode(), inplace=True)
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
columns_to_normilalize = ['Age', 'Fare']
df[columns_to_normilalize] = scaler.fit_transform(df[columns_to_normilalize])
df[columns_to_normilalize].head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_standardize = ['Age', 'Fare']
df[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])
df[columns_to_standardize].head()
```
